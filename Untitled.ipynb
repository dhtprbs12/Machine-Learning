```python
import tensorflow as tf
import numpy as np
from pandas.io.parsers import read_csv

#model = tf.global_variables_initializer() #initialize model
data = read_csv('price data.csv',sep=',') #read excel file separate by ','
xy = np.array(data, dtype=np.float64) #get data as row-col pattern into xy
print(xy)
```

```python
x_data = xy[:, 1:-1] #slicing grammar 
y_data = xy[:, [-1]] #get right most value using slicing grammar in Python (avgPrice)
print(y_data)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 4]) # X has 4 values
Y = tf.placeholder(tf.float32, shape=[None, 1]) # Y has 1 values
W = tf.Variable(tf.random_normal([4,1]), name="weight") #가중치 값 초기화
b = tf.Variable(tf.random_normal([1]), name="bias") #bias값 초기화
```

```python
hypothesis = tf.matmul(X,W) + b # initialize real hypothesis for machine learning
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.000005)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer()) #글로벌 변수 초기화
```

```python
for step in range(100001): # 10만회 학습
    cost_, hypo_, _ = sess.run([cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})
    if step % 500 == 0: # 500회 학습 마다
        print('#', step, "cost: ", cost_)
        print("Lettuce price: ", hypo_[0])
```

```python
saver = tf.train.Saver() #educated model save
save_path = saver.save(sess, "./saved.cpkt")
print("Succeeded to save")
```
